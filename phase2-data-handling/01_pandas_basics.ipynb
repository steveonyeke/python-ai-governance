```python
# First eval dataset with pandas
import time
from google import genai
from google.colab import userdata
import pandas as pd

#Setup
client = genai.Client(api_key=userdata.get('GOOGLE_API_KEY'))

# Function 1 - Ask the LLM
def ask_llm(prompt):
  response = client.models.generate_content(
      model="gemini-flash-latest",
      contents=prompt
  )
  return response.text

# Function 2 - Run eval and collect into a table
def run_eval(prompt_list, topic="General"):
  print(f"====== EVALUATION: {topic} ======")

  results = []

  for i, prompt in enumerate(prompt_list):
    response = ask_llm(prompt)
    word_count = len(response.split())
    results.append({
        "prompt_number": i + 1,
        "prompt": prompt,
        "response": response,
        "word_count": word_count
    })
    print(f"[{i+1}] Done - {word_count}words")
    time.sleep(2)

  # Covert to pandas table
  df = pd.DataFrame(results)
  print(f"\n====== EVAL COMPLETE ======")
  print(f"Total prompts: {len(df)}")
  print(f"Average response length: {df['word_count'].mean():.0f} words")
  print(f"Longest response: {df['word_count'].max()} words")
  print(f"Shortest response: {df['word_count'].min()} words")

  return df

# Prompts
ai_governance_prompts = [
    "What is the key perspective of EU AI ACT?",
    "What is the perspective of NIST AI framework?",
    "What is the need for RLHF in AI model evaluation?",
    "Why does AI governance matter for the future of humanity?",
]

# Run it
df = run_eval(ai_governance_prompts, topic="AI Governance Basics")

# Display the table
df
```

```python
# Save eval result to a file
df.to_csv("ai_governance_eval_results.csv", index=False)
print("Results saved to 'ai_governance_eval_results.csv'")

# Reload to verify it saved correctly
df_loaded = pd.read_csv("ai_governance_eval_results.csv")
print(f"Rows loaded: {len(df_loaded)}")
print(f"Columns: {list(df_loaded.columns)}")
```

```python
# Add quality signal columns to my existing dataframe
# Keywords we expect in good AI governance responses
governance_keywords = ["risk","safety","human","model","data","bias"]

#Check each response for each keyword
for keyword in governance_keywords:
  df[f"contains_{keyword}"] = df["response"].str.lower().str.contains(keyword)

# Add a score - how many keywords appear in each response
df["keyword_score"] = df[[f"contains_{keyword}" for keyword in governance_keywords]].sum(axis=1)

# Display results
print("====== KEYWORD ANALYSIS ======")
print(df[["prompt_number","prompt","word_count","keyword_score"]])
print(f"\nMax possible score: {len(governance_keywords)}")
print(f"Average keyword score: {df['keyword_score'].mean():.1f}")

#Save updated results
df.to_csv("ai_governance_eval_results.csv", index=False)
print("\nUpdated results saved to ai_governance_eval_results.csv")
```

```python
import time
import pandas as pd
from google import genai
from google.colab import userdata, drive

# Mount Google Drive - your permanent storage
drive.mount('/content/drive')

# Setup
client = genai.Client(api_key=userdata.get('GOOGLE_API_KEY'))

# Create a folder path in your Drive
SAVE_PATH = "/content/drive/MyDrive/python-ai-governance/data/"

# Create the folder if it doesn't exist
import os
os.makedirs(SAVE_PATH, exist_ok=True)

print("Google Drive mounted and ready.")
```

```python
def ask_llm(prompt):
    response = client.models.generate_content(
        model="gemini-flash-latest",
        contents=prompt
    )
    return response.text

def run_eval(prompt_list, topic="General"):
    print(f"====== EVALUATION: {topic} ======")
    results = []

    for i, prompt in enumerate(prompt_list):
        response = ask_llm(prompt)
        word_count = len(response.split())
        results.append({
            "prompt_number": i + 1,
            "prompt": prompt,
            "response": response,
            "word_count": word_count,
        })
        print(f"[{i+1}] Done — {word_count} words")
        time.sleep(2)

    df = pd.DataFrame(results)

    # Add keyword scoring
    governance_keywords = ["risk", "safety", "human", "model", "data", "governance"]
    for keyword in governance_keywords:
        df[f"contains_{keyword}"] = df["response"].str.lower().str.contains(keyword)
    df["keyword_score"] = df[[f"contains_{keyword}"
                               for keyword in governance_keywords]].sum(axis=1)

    # Save to Google Drive permanently
    filepath = SAVE_PATH + "ai_governance_eval_results.csv"
    df.to_csv(filepath, index=False)
    print(f"\nResults saved permanently to Google Drive")
    print(f"Path: {filepath}")

    return df, governance_keywords

# Run everything
ai_governance_prompts = [
    "What is the key perspective of EU AI ACT?",
    "What is the perspective of NIST AI framework?",
    "What is the need for RLHF in AI model evaluation?",
    "Why does AI governance matter for the future of humanity?",
]

df, governance_keywords = run_eval(ai_governance_prompts, topic="AI Governance Basics")
```

```python
# Find which keywords are missing per prompt
print("====== KEYWORD BREAKDOWN ======\n")

for i, row in df.iterrows():
  print(f"Prompt {row['prompt_number']}: {row['prompt'][:50]}...")
  for keyword in governance_keywords:
    status = "✅" if row[f"contains_{keyword}"] else "❌"
    print(f"  {status} {keyword}")
  print(f"  Score: {row['keyword_score']}/{len(governance_keywords)}\n")
```
