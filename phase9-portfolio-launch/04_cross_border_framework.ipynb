# Phase 9: Cross-Border Framework
**Goal**: Map whether Phase 8/9's EU AI Act and NIST deployer
      evidence transfers to Nigeria (NDPA 2023) and Canada
      (the original roadmap's third jurisdiction), as a thin
      layer on the existing 22-obligation scope.
**Note on Canada**: AIDA died on the order paper 6 January
      2025 and was never enacted. Canada's June 2026 national
      AI strategy chose voluntary certification over statutory
      obligations. As of June 2026, Canada is the only G7
      nation without a binding federal AI law, the honest
      finding for this jurisdiction.
**Regulatory mapping**: EU AI Act Articles 12, 14, 26, 27;
                    Nigeria NDPA 2023 Sections 25, 27, 37.
**Date**: June 2026.
**Status**: In Progress.

```python
import json
from datetime import datetime
from dataclasses import dataclass, field, asdict
from typing import Optional, List, Dict, Any

# - TRANSFER STATUS -
FULL_EQUIVALENT = "FULL_EQUIVALENT"
PARTIAL_EQUIVALENT = "PARTIAL_EQUIVALENT"
NO_EQUIVALENT = "NO_EQUIVALENT"

# - SCOPE NOTE -
# Afrispan's actual client is the deployer, the African SME using a
# third-party AI tool (OpenAI, Claude, a vendor hiring product), not
# the provider building a model from scratch. EU AI Act Article 26,
# NIST AI RMF, and ISO/IEC 42001 already cover deployer obligations
# well, which is why the 22-obligation Conformity Report scope built
# in 03_conformity_report.ipynb remains the primary scope here. This
# notebook adds a thin layer on top: for the obligations most directly
# tied to deployer duties, does the same evidence transfer to Nigeria
# (NDPA 2023, real binding AI-adjacent law, Afrispan's core African
# market) and Canada (the original roadmap's third jurisdiction).
#
# Note on Canada: the original 6-month roadmap specified Canada's AIDA
# (Bill C-27) as the third jurisdiction. AIDA died on the order paper
# on 6 January 2025 when Parliament was prorogued, and was never
# enacted. Canada's June 2026 national AI strategy explicitly declines
# to put any high-risk AI system on a statutory footing, choosing
# voluntary certification instead. As of June 2026, Canada is the only
# G7 nation without a binding federal AI law. This is the honest
# finding for this jurisdiction: not a partial transfer, but a
# transfer target that does not currently exist in binding form. The
# roadmap commitment is honored by including Canada and stating this
# plainly, rather than quietly substituting a different country.
#
# Phase 3-6 check: the same direct-verification standard used in
# 03_conformity_report.ipynb was applied here before writing this
# notebook. No Phase 3-6 evidence speaks to Articles 12, 26, or 27
# specifically. Phase 4's repeated "human review corrects automated
# classifier" finding is conceptually adjacent to human oversight
# obligations, but it is an evaluation-methodology finding, not
# deployed governance evidence, so it is not cited here as equivalent
# to Phase 8's kill-switch. This is stated rather than left silent.


@dataclass
class TransferNote:
  """One obligation's cross-border applicability to one jurisdiction."""
  base_obligation: str
  jurisdiction: str
  transfer_status: str
  local_provision: str
  gap_reasoning: str
  source_reference: str


def build_transfer_map(transfer_notes: List[TransferNote]) -> Dict[str, Any]:
  """Summarise a list of TransferNote into a structured cross-border report."""
  status_counts = {FULL_EQUIVALENT: 0, PARTIAL_EQUIVALENT: 0, NO_EQUIVALENT: 0}
  for note in transfer_notes:
    status_counts[note.transfer_status] += 1

  return {
      "generated_at": datetime.now().strftime("%Y-%m-%d %H:%M:%S"),
      "notes": [asdict(n) for n in transfer_notes],
      "status_counts": status_counts,
      "total_notes": len(transfer_notes),
  }


def render_transfer_markdown(transfer_map: Dict[str, Any]) -> str:
  badge = {
      FULL_EQUIVALENT: "[TRANSFERS FULLY]",
      PARTIAL_EQUIVALENT: "[TRANSFERS PARTIALLY]",
      NO_EQUIVALENT: "[NO LOCAL EQUIVALENT]",
  }

  lines = [
      "# Cross-Border Deployer Obligation Transfer Map",
      f"**Generated:** {transfer_map['generated_at']}",
      f"**Primary scope:** EU AI Act, NIST AI RMF, ISO/IEC 42001 "
      f"(see 03_conformity_report.ipynb, 22 obligations)",
      f"**Case-study jurisdictions:** Nigeria (NDPA 2023), Canada "
      f"(no binding federal AI law as of June 2026)",
      "", "---", "",
  ]

  current_obligation = None
  for note in transfer_map["notes"]:
    if note["base_obligation"] != current_obligation:
      current_obligation = note["base_obligation"]
      lines.append(f"## {current_obligation}")
      lines.append("")

    lines.append(f"### {note['jurisdiction']} {badge[note['transfer_status']]}")
    lines.append(f"\n**Local provision:** {note['local_provision']}")
    lines.append(f"\n{note['gap_reasoning']}")
    lines.append(f"\n*Source: {note['source_reference']}*")
    lines.append("")

  lines.append("---")
  counts = transfer_map["status_counts"]
  lines.append(f"\n**Summary: {counts[FULL_EQUIVALENT]} fully transfer, "
               f"{counts[PARTIAL_EQUIVALENT]} partially transfer, "
               f"{counts[NO_EQUIVALENT]} have no local equivalent.**")

  return "\n".join(lines)


# - THE TRANSFER MAP: REAL EVIDENCE, REAL LAW, FOUR OBLIGATIONS -
# Only obligations with a genuine, citable local-law difference are
# included. This is a thin layer, not a parallel compliance system.
print("====== CROSS-BORDER DEPLOYER OBLIGATION TRANSFER MAP ======\n")

transfer_notes = [

    # EU AI Act Article 14, Human Oversight + Phase 8 kill-switch evidence
    TransferNote(
        base_obligation="EU AI Act Article 14 (Human Oversight)",
        jurisdiction="Nigeria (NDPA 2023)",
        transfer_status=FULL_EQUIVALENT,
        local_provision="Section 37, right not to be subjected to a decision "
                         "based solely on automated processing where the "
                         "decision has legal or significant effect; includes "
                         "right to request human intervention.",
        gap_reasoning="Phase 8's kill-switch evidence (named human authorizer, "
                      "documented position before resuming) directly satisfies "
                      "the same underlying requirement Section 37 protects: a "
                      "human must meaningfully intervene before an automated "
                      "decision with significant effect stands. The legal "
                      "trigger differs (Section 37 attaches to the affected "
                      "individual's right; Article 14 attaches to the "
                      "deployer's design obligation), but the evidence an "
                      "Afrispan client would need to produce is the same.",
        source_reference="Nigeria Data Protection Act 2023, Section 37; "
                          "verified June 2026",
    ),
    TransferNote(
        base_obligation="EU AI Act Article 14 (Human Oversight)",
        jurisdiction="Canada (no binding federal AI law)",
        transfer_status=NO_EQUIVALENT,
        local_provision="No binding federal AI-specific human oversight "
                         "requirement exists. AIDA, which would have "
                         "established one, died on the order paper 6 "
                         "January 2025 and was never enacted.",
        gap_reasoning="There is no local provision for Phase 8's kill-switch "
                       "evidence to transfer to, since none currently exists "
                       "in binding form. Canada's June 2026 national AI "
                       "strategy explicitly chose voluntary certification "
                       "over statutory high-risk obligations, meaning this "
                       "gap is not an oversight in the law, it is the "
                       "deliberate current policy position. Ontario's Bill "
                       "194 imposes human-oversight-adjacent duties, but only "
                       "on public-sector bodies (hospitals, ministries, law "
                       "enforcement), not on private deployer SMEs, so it "
                       "does not change the picture for an Afrispan client.",
        source_reference="Montreal AI Ethics Institute, 'The Death of "
                          "Canada's Artificial Intelligence and Data Act'; "
                          "The Walrus, June 2026 national AI strategy "
                          "coverage; verified June 2026",
    ),

    # EU AI Act Article 26, Deployer Obligations, the core of this deliverable
    TransferNote(
        base_obligation="EU AI Act Article 26 (Deployer Obligations)",
        jurisdiction="Nigeria (NDPA 2023)",
        transfer_status=PARTIAL_EQUIVALENT,
        local_provision="Sections 25 (lawful basis), 27 (transparency about "
                         "automated decision-making), and the Data Protection "
                         "Impact Assessment requirement before high-risk "
                         "processing.",
        gap_reasoning="The general shape transfers: both regimes expect a "
                       "deployer to have a lawful basis, inform affected "
                       "people, and assess risk before high-risk processing. "
                       "What does not transfer is Article 26's specificity: "
                       "named individuals with documented competence and "
                       "authority, a minimum 6-month log retention period, "
                       "and a defined incident-reporting chain to a market "
                       "surveillance authority. The NDPA's DPIA requirement "
                       "is real and enforceable, but it does not specify the "
                       "same operational detail Article 26 does.",
        source_reference="Nigeria Data Protection Act 2023, Sections 25, 27; "
                          "verified June 2026",
    ),
    TransferNote(
        base_obligation="EU AI Act Article 26 (Deployer Obligations)",
        jurisdiction="Canada (no binding federal AI law)",
        transfer_status=NO_EQUIVALENT,
        local_provision="No deployer-specific AI obligation exists in "
                         "binding federal Canadian law as of June 2026.",
        gap_reasoning="This is the most significant finding in this "
                       "deliverable, and it is stronger than a partial gap: "
                       "Canada has no binding federal AI statute at all to "
                       "compare against. AIDA would have introduced exactly "
                       "this kind of deployer obligation (design, "
                       "development, deployment duties for high-impact "
                       "systems), but it died with Bill C-27. PIPEDA "
                       "provides general privacy protection, and OSFI "
                       "Guideline E-23 covers model risk management for "
                       "federally regulated financial institutions "
                       "specifically, but neither imposes the named-oversight, "
                       "logging, and monitoring package Article 26 does for "
                       "AI deployers generally. As of June 2026, Canada is "
                       "the only G7 nation without a binding federal AI law.",
        source_reference="AIMag, 'Canada's AI Regulation Gap Widens as AIDA "
                          "Dies in Parliament'; BLG, 'A Turning Point for AI "
                          "in Canada in 2026'; verified June 2026",
    ),

    # EU AI Act Article 12, Record-Keeping + Phase 8 immutable audit log
    TransferNote(
        base_obligation="EU AI Act Article 12 (Record-Keeping)",
        jurisdiction="Nigeria (NDPA 2023)",
        transfer_status=PARTIAL_EQUIVALENT,
        local_provision="General Application and Implementation Directive "
                         "(GAID) 2025 requires periodic compliance audits "
                         "and an annual Compliance Audit Return for "
                         "data controllers/processors of major importance.",
        gap_reasoning="Phase 8's hash-chained immutable audit log exceeds "
                       "what Nigerian law actually requires. The GAID asks "
                       "for periodic audits and annual filings, not a "
                       "tamper-evident technical record. An Afrispan client "
                       "meeting Article 12's standard is, in this respect, "
                       "already over-delivering relative to the binding "
                       "local requirement, which is a genuine selling point, "
                       "not a gap.",
        source_reference="NDPA General Application and Implementation "
                          "Directive 2025; verified June 2026",
    ),
    TransferNote(
        base_obligation="EU AI Act Article 12 (Record-Keeping)",
        jurisdiction="Canada (no binding federal AI law)",
        transfer_status=NO_EQUIVALENT,
        local_provision="No AI-specific record-keeping requirement exists "
                         "in binding federal Canadian law.",
        gap_reasoning="As with Article 26, there is no local AI-specific "
                       "provision to compare against. PIPEDA's general "
                       "accountability principle implies some form of "
                       "record-keeping for personal data processing, but "
                       "nothing approaching Article 12's logging "
                       "specificity exists for AI systems generally. Phase "
                       "8's immutable audit log remains good practice and "
                       "the kind of evidence a future Canadian AI statute, "
                       "if one is tabled within the 18 to 24 month window "
                       "some commentators expect, would likely require.",
        source_reference="AIMag, 'Canada's AI Regulation Gap Widens as AIDA "
                          "Dies in Parliament'; verified June 2026",
    ),

    # EU AI Act Article 27, FRIA, no equivalent anywhere checked
    TransferNote(
        base_obligation="EU AI Act Article 27 (Fundamental Rights Impact Assessment)",
        jurisdiction="Nigeria (NDPA 2023)",
        transfer_status=NO_EQUIVALENT,
        local_provision="No fundamental-rights-specific impact assessment "
                         "requirement exists; the NDPA's DPIA is a data "
                         "protection impact assessment, narrower in scope.",
        gap_reasoning="A DPIA under the NDPA asks whether processing creates "
                       "a high risk to data subjects' privacy. A FRIA under "
                       "Article 27 asks a broader question: does this system "
                       "affect fundamental rights beyond privacy, including "
                       "non-discrimination, human dignity, and effective "
                       "remedy. The Phase 9 FRIA generator's four-element "
                       "structure (named individual, timestamp, "
                       "pre-commitment rationale, structured interrogation) "
                       "is not required anywhere in Nigerian law, but it is "
                       "a genuine differentiator Afrispan can offer clients "
                       "who want assurance beyond the binding minimum.",
        source_reference="Nigeria Data Protection Act 2023; verified June 2026",
    ),
    TransferNote(
        base_obligation="EU AI Act Article 27 (Fundamental Rights Impact Assessment)",
        jurisdiction="Canada (no binding federal AI law)",
        transfer_status=NO_EQUIVALENT,
        local_provision="No equivalent exists in binding federal Canadian "
                         "law. AIDA's 2023 amendments would have introduced "
                         "impact assessment requirements for high-impact "
                         "systems, but AIDA was never enacted.",
        gap_reasoning="This is the same structural gap as Article 26 and "
                       "Article 12: the instrument that would have created "
                       "this obligation died with Bill C-27. The federal "
                       "Treasury Board's Algorithmic Impact Assessment Tool "
                       "exists, but it applies only to federal government "
                       "institutions' own automated decision systems, not "
                       "to private-sector deployers. Afrispan's FRIA "
                       "template is, for a Canadian client today, pure "
                       "differentiation: nothing in binding law requires it, "
                       "but it is the strongest evidence available if "
                       "federal legislation is tabled in the 18 to 24 month "
                       "window some commentators expect.",
        source_reference="Schwartz Reisman Institute, 'What's Next After "
                          "AIDA?'; verified June 2026",
    ),
]

transfer_map = build_transfer_map(transfer_notes)
print(render_transfer_markdown(transfer_map))
```

### Key Findings

1. **Nigeria's NDPA Section 37 is the one full transfer.** Phase 8's
   kill-switch evidence, a named human authorizer documenting a
   substantive position before resuming, satisfies the same underlying
   requirement Section 37 protects, even though the legal trigger
   differs (individual right vs deployer design obligation).

2. **Phase 8's immutable audit log exceeds what Nigerian law actually
   requires.** The GAID 2025 asks for periodic audits and annual
   filings, not a tamper-evident technical record. This is a genuine
   selling point for Afrispan: clients meeting the EU AI Act standard
   are already over-delivering relative to Nigeria's binding floor.

3. **Canada's gap is not narrow, it is total.** AIDA, the instrument
   that would have created deployer obligations, named oversight,
   record-keeping, and impact assessment requirements, died on the
   order paper 6 January 2025 and was never enacted. Canada's own June
   2026 national AI strategy explicitly chose voluntary certification
   over statutory obligations. As of June 2026, Canada is the only G7
   nation without a binding federal AI law. Every Canada transfer note
   is NO EQUIVALENT not because the comparison is unfavorable, but
   because there is currently nothing binding to compare against.

4. **Article 27 (FRIA) has no equivalent in either jurisdiction
   checked.** Nigeria's DPIA is privacy-scoped, narrower than Article
   27's fundamental-rights breadth. Canada's federal Algorithmic Impact
   Assessment Tool applies only to government institutions, not
   private deployers. Afrispan's FRIA template is, for clients in both
   markets today, pure differentiation: nothing in binding law requires
   it, but it is the strongest evidence available if either
   jurisdiction's regulatory landscape changes.
